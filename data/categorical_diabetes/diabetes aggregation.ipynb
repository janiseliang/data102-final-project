```python
import pandas as pd
import os
```

## Combines various diabetes data tables into a single csv

data pulled from https://gis.cdc.gov/grasp/diabetes/diabetesatlas-surveillance.html

```python
folder = "./"
os.listdir(folder)
```

```python
data = None
for fname in os.listdir(folder):
  if '.csv' in fname:
    data_name = '-'.join(fname.split('.')[0].split('_')[1:])
    df = pd.read_csv(folder+fname, header=2, index_col=0)
    df = df.iloc[1:, 0:1].rename(columns={'Percentage': data_name})
    if data is None: data = df
    else: data = data.join(df)
data.head()
```

```python
data.to_csv("../categorical_diabetes.csv")
```
